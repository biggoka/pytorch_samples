```python
import pandas as pd
import numpy as np
import quandl
import datetime
```

```python
tickers = ['FB', 'MSFT', 'AMZN', 'TSLA', 'AMD', 'QCOM']
```

```python
start = datetime.datetime(2002,1,1)
end = datetime.date.today()

stock_data = []

for ticker in tickers:
    stock_df = quandl.get("WIKI/" + ticker, start_date=start, end_date=end)
    stock_df = stock_df[['Open', 'High', 'Low', 'Close', 'Volume']]
    stock_df.reset_index(inplace=True)
    stock_df.columns = ['date', 'open', 'high', 'low', 'close', 'volume']
    stock_df['ticker'] = ticker
    stock_data.append(stock_df.copy())
    
stock_data = pd.concat(stock_data, axis=0)
```
